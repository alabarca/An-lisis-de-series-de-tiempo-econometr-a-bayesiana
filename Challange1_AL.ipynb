```python
# import libraries

import pandas as pd
import numpy as np
pd.set_option('display.max_columns', 100)
import matplotlib.pyplot as plt
import seaborn as sns
import time
from datetime import datetime
import re

import twint
import nest_asyncio
from ast import literal_eval
```

```python
# define search parameters

keywords = ['#IranRevolution2022 OR #StopExecutionsInIran OR #SoniaSharifi']
search_start_date = '2022-12-08'
number_of_tweets_to_scrape = 1000
```

```python
# Scrape

nest_asyncio.apply()
c = twint.Config()

# search params
c.Search = keywords
c.Since = search_start_date
c.Limit = number_of_tweets_to_scrape

# save
c.Store_csv = True
c.Output = 'account_tweets.csv'
c.Pandas = True

# run scrape
twint.run.Search(c)
```

```python
# view the scraped tweets
tweets_by_keyword = twint.storage.panda.Tweets_df

# convert dates to datetime object 
tweets_by_keyword.date = pd.to_datetime(tweets_by_keyword.date)
tweets_by_keyword['date_only'] = tweets_by_keyword.date.dt.date
```

```python
####Explore the text
###I choosed the text more used by the users and displayed the link of any tweet that contain the first text: "#StopExecutionsInIran"


df2 = tweets_by_keyword.groupby(['tweet'])[['id']].count()   
df2.sort_values(by=['id'], ascending=False).head(10)

```

```python
##Identify the tweet with most likes

most_liked_tweet_from_keywords = tweets_by_keyword.loc[tweets_by_keyword.nlikes == tweets_by_keyword.nlikes.max()]
print('Most liked Tweet:')
most_liked_tweet_from_keywords
most_liked_tweet_from_keywords['tweet'].item()
```

```python
##Find the 10 most used languages among all scraped tweets
df3 = tweets_by_keyword.groupby(['language'])[['id']].count()   
df3.sort_values(by=['id'], ascending=False).head(10)
```

```python
##Find the 10 most frequent hastags and show them in a bar plot
tweets_by_keyword['hashtags'] = tweets_by_keyword['hashtags'].astype('string')
df4=tweets_by_keyword
df4['hashtags'] = df4['hashtags'].apply(literal_eval)
hashtags = df4.explode('hashtags')
hashtags = hashtags.groupby(['hashtags'])[['id']].count()   
hashtags.sort_values(by=['id'], ascending=False).iloc[0:10,:]
```
